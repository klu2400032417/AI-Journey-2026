```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2023,2024,2025,2026])
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])
y3=np.array([13,15,20,30])
plt.title("Class size",fontsize=20,
         family="Arial",fontweight="bold",color="purple")
plt.xlabel("Year", fontsize=20,
           family="Arial",fontweight="bold",color="blue")
plt.ylabel("students", fontsize=20,
           family="Arial",fontweight="bold",color="blue")
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
# ticks make the x axis same as the list does not add 0.5 extra by using ticks
plt.xticks(x)
#  changes the x and y axis number color to blue
plt.tick_params(axis="both",colors="blue")
plt.show()
```
