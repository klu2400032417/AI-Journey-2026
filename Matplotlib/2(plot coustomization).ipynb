```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([2023,2024,2025,2026])
y=np.array([15,25,30,20])
plt.plot(x,y,marker=".",markersize=30,markerfacecolor="yellow",
         markeredgecolor="pink",linestyle="dashed"
        ,linewidth=3,color="red")
#  in short markerfacecolor u can write as mfc
# markeredgecolor means circle color
#  dashed or dotted or dashdot or /none or solid etc...
plt.show()

```

```python
y1=np.array([15,25,30,20])
y2=np.array([17,23,38,5])
plt.plot(x,y1,marker=".",markersize=30,markerfacecolor="yellow",
         markeredgecolor="pink",linestyle="dashed"
        ,linewidth=3,color="red")
plt.plot(x,y2,marker=".",markersize=30,markerfacecolor="yellow",
         markeredgecolor="pink",linestyle="dashed"
        ,linewidth=3,color="red")
plt.show()
```

```python
line_style=dict(
    marker=".",markersize=30,markerfacecolor="yellow",
         markeredgecolor="pink",linestyle="dashed"
        ,linewidth=3
)
# **means unpacking our dictionary which contains the styles

plt.plot(x,y1,color="red",**line_style)
plt.plot(x,y2,color="orange",**line_style)
plt.show()
```
