```python
import matplotlib.pyplot as plt
import numpy as np
# Scatter graphs= Shows the relationship btw 2 variables
#                Help to identify a coorelation (+,-,None)
#                Example: Study hours vs Test scorces

x=np.array([0,1,1,2,3,4,5,6,7,7,8])  # Hours studied
y=np.array([55,60,65,62,68,70,75,78,82,85,87])# Grades
plt.xlabel("Hours studied")
#  alpha means transperency of the dots and s is for size
plt.ylabel("Grade")
plt.title("test scorces")
plt.scatter(x,y,color="yellow",
          alpha=0.5,s=100,label="ClassA")
# legend is used to mention the label
plt.legend()
plt.show()
```
