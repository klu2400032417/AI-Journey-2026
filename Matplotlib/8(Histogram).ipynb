```python
import matplotlib.pyplot as plt
import numpy as np
# Histogram = A visual representation of the distribution of quantitive data
#             They grp values into bins (intervals)
#             and counts how many fall in each range

scorces=np.random.normal(loc=80, scale=10,size=100)
# loc is  meadian of the data center point scale means the 
# amount of deviation we need from the median
scorces=np.clip(scorces,0,100)
#  for any scorces below 0 is clip them to 0 and max is clipped to 100 means above 100 
# is setted to 100
plt.title("exam scorces")

plt.hist(scorces,bins=10,color="yellow",edgecolor="Red")
# bins means intervals in the graph
plt.show()
```
