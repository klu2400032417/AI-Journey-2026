```python
import matplotlib.pyplot as plt
import numpy as np
# Bar Chart = compare categories of data by representing each category with a bar
categories=np.array(["Grains","Fruits","vegetables","Protein","Diary","sweets"])
values=np.array([4,3,2,5,3,1])
plt.bar(categories,values,color="yellow")
#  barh is for horizontal bar chart 
# plt.barh(categories,values,color="yellow")
plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantiy")
plt.show()
```
