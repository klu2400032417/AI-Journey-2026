```python
# Scalar arethematic
import numpy as np
arr=np.array([1,2,3])
print(arr+1)
print(arr*2)
print(arr/3)
print(arr **2)
```

```python
# Vectorized math functions

print(np.sqrt(arr))
a=np.array([1.29,3.22])
print(np.round(a))
print(np.floor(a))
print(np.ceil(a))
print(np.pi)
```

```python
radii=np.array([1,2,3])
print(np.pi* radii **2)
# pi r square area of circle
```

```python
#element wise arthematic
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1**arr2)
```

```python
# comparison operators
scorces=np.array([91,55,100,73,82,64])
print(scorces==100)
print(scorces>=60)
# very intresting
scorces[scorces<60]=0
print(scorces)
```
