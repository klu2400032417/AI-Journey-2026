```python
import numpy as np
random_number_generator=np.random.default_rng()
# np.random.default_rng(seed=1) here seed represents the generated number should be repeated again(to reproduce same results)
print(random_number_generator.integers(low=1,high=7,size=3))
#  mentioning low and high words is not necessary optional thing
# in size we can use dimensions like size=(3,2) which means 3 rows and 2 col
```

```python
print(np.random.uniform(low=-1,high=1))
# uniform means uniform distribution
```

```python
np.random.seed(seed=1)
print(np.random.uniform(low=-1,high=1))
```

```python
rng=np.random.default_rng()
array=np.array([1,2,3,4,5])
rng.shuffle(array)
print(array)
```

```python

fruits=np.array(['apple','banana','grapes'])
fru=rng.choice(fruits,size=(3,2))
print(fru)
```
