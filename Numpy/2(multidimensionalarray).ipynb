```python
import numpy as np
array=np.array('A')
print(array.ndim)
# ndim means number of dimensions
```

```python
arr=np.array(['a','b','c'])

ar=np.array([['a','b','c'],['d','e','f']] )

a=np.array([[['a','b','c'],['d','e','f']],
           [['a','b','c'],['d','e','f']]])
print(arr.ndim)
print(ar.ndim)
print(a.ndim)
print(a.shape)# show depth, number of rows and number of  columns
print(a[0][0][0])# chain indexing
print(a[0,0,0])# this is multi dimensional indexing which is faster than chain indexing
```

```python
word=a[1,1,0]+a[0,1,1]
# this + is concatenation
print(word)
```
