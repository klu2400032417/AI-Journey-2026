```python
# Pandas means panel data works with series and data frames
# series means 1 dimensional labeled column and dataframe means 2 dimensional labeled table
# we import, dispaly, manipulate, export data

import pandas as pd

print(pd.__version__)
```

```python
# Series =A pandas 1 dimensional labeled array that canhold any data type 
# Think of it like a single column in a spreadsheet (1d)
data=[100,200,300]
series=pd.Series(data)
print(series)
```

```python
data=[100,200,300]
series=pd.Series(data, index=['a','b','c'])
print(series)
```

```python
series.loc['c']=102
print(series.loc['a'])
print(series.loc['c'])
# loc means location by variable it returns the value at that variable
```

```python
print(series.iloc[0])
# iloc means integer location
```

```python
data=[100,102,104,200,102]
series=pd.Series(data,index=['a','b','c','d','e'])
print(series[series<=160])
```

```python
calories={
    "Day1":1750,
    "Day2":2100,
    "Day3":1700
}
series=pd.Series(calories)
print(series)
print(series.loc["Day3"])
series.loc["Day3"]+=500
print(series.iloc[0])
```
