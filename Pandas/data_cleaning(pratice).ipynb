```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name":["A","B","C","C"],
    "Marks":[90,np.nan,75,75]
})
print(df.isnull())
```

```python
df=df.fillna({"Marks":0})
print(df)
```

```python
df=df.drop_duplicates("Name")
print(df)
```

```python
df["Name"]=df["Name"].astype(str)
df["Name"]=df["Name"].replace({'C':'B'})
print(df)
```

```python
print(df.duplicated(subset=["Name"]))
```
