```python
import pandas as pd
# DataFrame=A tabular data strucutre with rowsand columns. (2d)
# similar toexcel spread sheet
data={
    "Name":["SpongeBob","Patrick","Squidward"],
    "Age":[30,35,50],
}
df=pd.DataFrame(data)
print(df)
```

```python
df=pd.DataFrame(data,index=["Employee1","Employee2","Employee3"])
print(df)
```

```python
print(df.loc["Employee1"])
print(df.iloc[0])
```

```python
# Add a new column
df["Job"]=["Cook","N/A","Cashier"]
print(df)
```

```python
# Add a new row
new_row=pd.DataFrame([{"Name":"Sandy","Age":28,"Job":"Engineer"},
                      {"Name":"asha","Age":60,"Job":"Manager"}]
                     ,index=["Employee4","Emploee5"])
df=pd.concat([df,new_row])
print(df)
```
