```python
# Data Cleaning =the process of removing/fixing:
#              incomplete, incorrect, or irrelevant data.
#              ~ 75% of work done with pandas is data cleaning
import pandas as pd
df=pd.read_csv("train.csv")
# 1.Drop irrelevant columns
df=df.drop(columns=["s12"])
print(df)
```

```python
# 2.Handling missing data
df=df.dropna(subset=["s13"])
# in s13 col drops the values which r not having any value
#  na means not available
print(df)
```

```python
df=df.fillna({"s13":"None"})
# replaces the missing values with none
print(df)
```

```python
# 3.Fix inconsitent values
df["s13"]=df["s13"].replace({1:"ONE"})
print(df["s13"])
```

```python
# 4.Standardize text
df["s13"]=df["s13"].str.lower()
print(df["s13"])
```

```python
# 5.Fix data types
df["s13"]=df["s13"].astype(String)
# 6.Remove duplicate values
df=df.drop_duplicates()
```
