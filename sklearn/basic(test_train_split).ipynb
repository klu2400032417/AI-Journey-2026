```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split

X, y = fetch_california_housing(return_X_y=True)

print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    #  means 80% data is for training and 20% is for testing
    random_state=42
    # "Use 42 as the starting seed so that the random 
    # split is the same every time I run the code."
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```
